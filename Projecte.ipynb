```python
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np
from datetime import datetime
import seaborn as sns
```

```python
embassaments_df = pd.read_csv('embassaments.csv')
embassaments_df.head()
```

```python
Sau_df = embassaments_df.loc[embassaments_df['Estació'] == 'Embassament de Sau (Vilanova de Sau)'].copy()
```

```python
Sau_df.head()
```

```python
Sau_df.info()
```

```python
Sau_df['Dia'] = pd.to_datetime(Sau_df['Dia'], format ='%d/%m/%y')
```

```python
Sau_df.info()
```

```python
Sau_df.head()
```

```python
Sau_df_sorted = Sau_df.sort_values(["Dia"]).copy()
```

```python
Sau_df_sorted
```

```python
fig = plt.figure(figsize=(30, 20))

ax = fig.add_subplot(111)

ax.set_xticklabels(Sau_df_sorted['Dia'], fontsize=20, rotation=45)
ax.set_yticklabels(Sau_df_sorted['Volum embassat (hm3)'], fontsize=20)
ax.set_xlim(datetime(2000,1,1), datetime(2022,12,13))

ax.plot(Sau_df_sorted['Dia'] , Sau_df_sorted['Volum embassat (hm3)']  , label = ' capacitat ', linewidth = 3 )

plt.xlabel('Día', fontsize=20) 
plt.ylabel('Volum embassat (hm3)', fontsize=20) 
  
plt.title("Volum aigua Pantà Sau del 2000 al 2022", fontsize=25)
```

```python
sns.lineplot(x = "Dia", y = "Volum embassat (hm3)", data=Sau_df_sorted)
```

```python
ts = Sau_df_sorted.set_index('Dia')
ts
```

```python
ts['Dia'] = ts.index
```

```python
ts
```

```python
ts_22 = ts['2022-01-01':'2022-12-13']
```

```python
ts_22
```

```python
ts_22.index = range(len(ts_22.index))
```

```python
ts_22
```

```python
ts_22.info()
```

```python
fig = plt.figure(figsize=(30, 22))

# Adds subplot on position 1
ax = fig.add_subplot(221)
ax.set_xlim(datetime(2022,1,1), datetime(2022,12,13))
# Adds subplot on position 2
ax2 = fig.add_subplot(223)
ax2.set_xlim(datetime(2022,1,1), datetime(2022,12,13))

ax.plot(ts_22['Dia'] , ts_22['Volum embassat (hm3)']  ,  linewidth = 3 )
ax2.plot(ts_22['Dia'] , ts_22['Percentatge volum embassat (%)']  ,  linewidth = 3 )
plt.show()
```
